### Count and add new columns to the table

Let's add variables to the table, namely

    price_per_meter - price per square metre
    weekday_publication - weekday of publication
    month_publication - month of publication 
    year_publication - year of publication 
    floor_category - type of flat floor 
    city_centers_nearest_km - proximity to city centre

<strong> Adding the price per square metre </strong>

```python
data['price_per_meter'] = data['last_price'] / data['total_area']
data.head()
```

<strong> Adding month, day of week and year column </strong>

```python
data['weekday_publication'] = data['first_day_exposition'].dt.weekday
data['month_publication'] = data['first_day_exposition'].dt.month
data['year_publication'] = data['first_day_exposition'].dt.year
data.head(10)
```

<strong> Adding column categorisation by floor </strong>

```python
def put_floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
data['floor_category'] = data.apply(put_floor_category, axis=1)
data.head(10)
```

```python
data['city_centers_nearest_km'] = round(data['city_centers_nearest'] / 1000)
data.head(10)
```

```python
data['city_centers_nearest'].describe()
```
